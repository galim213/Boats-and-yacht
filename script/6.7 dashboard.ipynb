```python
groupedby_type = df.groupby('boat_type')['views_last_7_days'].sum().sort_values(ascending=False)
top_5_boats = groupedby_type.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_boats.index, y=top_5_boats.values, palette="deep")
plt.title("Top 5 Boats by Views in Last 7 Days")
plt.xlabel("Boat Type")
plt.ylabel("Total Views")
plt.xticks(rotation=45)
plt.show()
```

```python

# create a set for ploting data country views
data_to_plot = df[['country','views_last_7_days']]
data_to_plot.head()
```

```python

groupedby_country = df.groupby('country')['views_last_7_days'].sum().sort_values(ascending=False)
top_5_countries = groupedby_country.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_countries.index, y=top_5_countries.values, palette="deep")
plt.title("Top 5 Countries by Total Views in Last 7 Days")
plt.xlabel("Country")
plt.ylabel("Total Views")
plt.xticks(rotation=45)
plt.show()
```

```python
groupedby_type = df.groupby('material')['views_last_7_days'].sum().sort_values(ascending=False)
top_5_material = groupedby_type.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_material.index, y=top_5_material.values, palette="deep")
plt.title("Top 5 Material by Views in Last 7 Days")
plt.xlabel("Material")
plt.ylabel("Total Views")
plt.xticks(rotation=45)
plt.show()
```
